```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('flexural.csv')
```

```python
df.head()
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.ensemble import AdaBoostRegressor
```

```python
from sklearn.model_selection import train_test_split,GridSearchCV
```

```python
X=df.drop('f',axis=1)
y=df['f']
```

```python
X_train, X_test, y_train, y_test = train_test_split( X, y, test_size=0.15, random_state=101)
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score
```

```python
def run_model(model,X_train,y_train,X_test,y_test):
    
    model.fit(X_train,y_train)
    
    preds=model.predict(X_test)
    
    rmse=mean_squared_error(y_test,preds)**0.5
    
    r2=r2_score(y_test,preds)
    
    print(f'Root Mean Squared Error: {rmse}')
    
    print(f'r2 score: {r2}')
```

```python
RD=RandomForestRegressor(random_state=101)
```

```python
run_model(RD,X_train,y_train,X_test,y_test)
```

```python
help(RandomForestRegressor)
```

```python
n_estimators=[64,100,128,200]
max_features= [2,3,4]
bootstrap = [True,False]
oob_score = [True,False]
```

```python
param_grid = {'n_estimators':n_estimators,
             'max_features':max_features,
             'bootstrap':bootstrap,
             'oob_score':oob_score} 
```

```python
rfc = RandomForestRegressor(random_state=101)
grid = GridSearchCV(rfc,param_grid)
```

```python
run_model(grid,X_train,y_train,X_test,y_test)
```

```python
grid.best_params_
```

```python
ada=AdaBoostRegressor(random_state=101)
```

```python
n_value=[50,100,150,200]

for n in n_value:
    model=AdaBoostRegressor(n_estimators=n,random_state=101)
    run_model(model,X_train,y_train,X_test,y_test)
```
